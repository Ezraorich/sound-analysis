```python
#3 
choice =''
while choice !=0:
    print ('Choose conversion\n1: F to C or \n2: C to F')
    choice= int(input())
    
    if choice == 1:
        print ('Enter the temperature in Fahrenheit:\n')
        b=int(input())
        print (b,'F= ',(b-32)*5/9, 'C\n')
    elif choice ==2:
        print ('Enter the temperature in Celcius:\n')
        b=int(input())
        print (b,'C= ',(b*9/5)+32,'F\n')
    else:
        print ('the end. user quits')
        break
```
