```python
from scipy.stats import mode

Mode = []
Mode2 = []

for i in keys:
    m = true_test_values[true_test_values['key']==i]['correlation'].values
    m2 = true_test_values[true_test_values['key']==i]['predicted values'].values
    
    #print(m)
    
    s = mode(m)[0]
    s2 = mode(m2)[0]
    
    Mode.append(s[0])
    Mode2.append(s2[0])
```
