```python
def plot_spectrogram(Y, sr, hop_length, y_axis ='linear'):
    plt.figure(figsize = (25,10))
    librosa.display.specshow(Y, 
                              sr =sr,
                             hop_length = hop_length,
                             x_axis ='time',
                             y_axis = y_axis)
    plt.colorbar(format = '\%+2.f')
    plt.savefig('1.png')
```

```python
plot_spectrogram(Y_log_scale, sr, hop_size, y_axis = 'log')
```

```python
data_dir ='D:/COVID_COUGH_SOUNDS/50ms_overlap_458_aug_balanced/50S_overlap_aug_balanced_458/test/pneumonia'
audio_files =glob(data_dir + '/*.wav')
print(len(audio_files))
stft_path= 'D:/COVID_COUGH_SOUNDS/50ms_overlap_458_aug_balanced/stft 50S overlap chunks aug 458/test/pneumonia/'
```

```python
%%time

frame_size = 2048
hop_size = 512
for file in range(0, len(audio_files), 1):
    count=0
    
    audio, sfreq =lr.load(audio_files[file])
    
    
    

    #scale, sr = librosa.load('D:/COVID_COUGH_SOUNDS/50S_overlap_458/50msOverlap_chunks_notAUG/test/allergies/1591363379163_0.wav')
    
    S_scale  = librosa.stft(audio, n_fft = frame_size, hop_length = hop_size)
    Y_scale = np.abs(S_scale)**2
    Y_log_scale = librosa.power_to_db(Y_scale)
    
    plt.figure(figsize = (25,10))
    librosa.display.specshow(Y_log_scale, sr =sfreq,
                        hop_length = 512,
                        x_axis ='time',
                        y_axis = 'log')
    plt.colorbar(format = '\%+2.f')
    name = get_sound_id_from_file_path(audio_files[file])
    plt.savefig(stft_path+str(name)+'.png')
    plt.clf()
    
    
    
    
    count+=1
print(count)
```
